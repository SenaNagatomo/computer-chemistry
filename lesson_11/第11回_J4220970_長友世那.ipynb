```python
import numpy as np
from sklearn.linear_model import LinearRegression

x = 2 * np.random.rand(100, 1)
y =  4 + 3 *  x  + np.random.randn(100, 1)

model = LinearRegression()
model.fit(x, y)

coeff = model.coef_
intercept = model.intercept_
print("係数:", coeff)
print("切片:", intercept)
```

```python
predict_from_x = model.predict(x)

z = 2 * np.random.rand(100, 1)
predict_from_z = model.predict(z)

import matplotlib.pyplot as plt
plt.scatter(x, y, color='blue', label='sample point')
plt.plot(x, predict_from_x, color='red', label='predicted value')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

#  課題１③
考察：　①の学習モデルは散布図からして上手くできた。次に、①の学習モデルを使ってdata2に対するtarget2の予測を行ったが、
散布図は相関関係を示さなかった。よって、１と２の間には、相関関係はないと言える。

```python
# 課題2
import numpy as np
from sklearn.cross_decomposition import PLSRegression

pls = PLSRegression(n_components=2)
pls.fit(data2, target2)
predict_from_data2_2 = pls.predict(data2)

import matplotlib.pyplot as plt
import japanize_matplotlib

plt.scatter(target2, predict_from_data2_2, color='blue', label='sample point')
plt.xlabel('実測値')
plt.ylabel('予測値')
plt.legend()
plt.show()
```

```python
# 課題3
import numpy as np
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=500, random_state=42)
rf.fit(data2, target2)
predict_from_data2_3 = rf.predict(data2)

plt.scatter(target2, predict_from_data2_3, color='red', label='sample point')
plt.scatter(target2, predict_from_data2_2, color='blue', label='sample point')
plt.xlabel('実測値')
plt.ylabel('予測値')
plt.legend()
plt.show()
```
